- Specify the utility function: constant relative risk aversion, CRRA-type utility function

$$
  u(c) = \frac{c^{1-\gamma}-1}{1-\gamma}
$$

- $1/\gamma$is the intertemporal elasticity of substitution

- When $\gamma \rightarrow 1$, $u(c)=\ln(c)$ holds due to the L'Hopital's rule

- The utility function here is ordinal (there is no meaning in its level or amount itself, but there is in the order suggested by its amount) and it is no problem with $u(c)=\frac{c^{1-\gamma}}{1-\gamma}$!

- The intertemporal elasticity of substitution is defined as follows:

$$
  \frac{d\ln(c_{2}/c_{1})}{dr}
$$

- From the Euler equation,

\begin{align*}
  & 1 = (1+r)\beta(c_{2}/c_{1})^{-\gamma} \\
  \Leftrightarrow & 0 = r-\rho -\gamma\ln(c_{2}/c_{1}) \\
  \Leftrightarrow & \gamma\ln(c_{2}/c_{1}) = r-\rho \\
  & \therefore \frac{d\ln(c_{2}/c_{1})}{dr} = \frac{1}{\gamma} 
\end{align*}

where $\ln(1+r) \approx r, \beta = (1+\rho)^{-1} \Leftrightarrow \ln\beta \approx -\rho $

- The L'Hopital's rule: When $f(a)=g(a)=0$,

$$
  \lim_{x \rightarrow a}\frac{f(x)}{g(x)}=\lim_{x \rightarrow a}\frac{f'(x)}{g'(x)}
$$

- Let $f(\gamma)=c^{1-\gamma}-1=\big[\exp(\ln(c))\big]^{1-\gamma}-1=\exp\big[(1-\gamma)\ln(c)\big]-1, g(\gamma)=1-\gamma$ then

\begin{align*}
  & \lim_{\gamma \rightarrow 1}\frac{f(\gamma)}{g(\gamma)} \\
  =& \lim_{\gamma \rightarrow 1}\frac{f'(\gamma)}{g'(\gamma)} \\
  =& \lim_{\gamma \rightarrow 1}\frac{\exp\big[(1-\gamma)\ln(c)\big]\big(-\ln(c)\big)}{-1} = \ln(c)
\end{align*}


```julia
function CRRA(cons::Float64, γ::Float64)
    if γ != 1.0
       return (cons^(1.0-γ)-1) / (1.0-γ) 
    else
       return log(cons)
    end
end
```

```julia
using Plots

n = 1001
gamvec = collect(LinRange(0.9,1.1,n))
uvec = zeros(n)
cons = 2.0

for i = 1:n

    γ = gamvec[i]
    uvec[i] = CRRA(cons,γ)
    
end

plot(gamvec,uvec)
```
