# Local approximation and global approximation
<!-- # 局所的近似と大域的近似 -->


- The standard solution method for dynamic macroeconomic models: Dynamic programming
<!-- - 動学的マクロ経済モデルの代表的な解法：動的計画法(dynamic programming) -->


- The current macroeconomic models are based on the neoclassical growth model
  - The dynamic optimization problem assuming that the representative household and firm will do economic activity for infinite future
  - The baseline model for real business cycle models and New Keynesian DSGE models
<!-- - 現在のマクロ経済モデルは新古典派成長モデル(neoclassical growth model)に支えられている -->
<!--   - 代表的個人（家計）と代表的企業が経済活動を無限期間行うと想定した際の動学的最適化問題 -->
<!--   - 実物的景気循環理論やニューケインジアンDSGEモデルなどのベース -->

- For similar economic problems, there are different solution methods corresponding to the model characteristics and what to know
<!-- - 似たような経済問題であっても、モデルの特性や知りたいことに応じて解法はいくつか存在 -->


- Two approaches
  - One to solve the model around the steady state locally
  - One to solve the model globally
<!-- - 2つのアプローチ -->
<!--   - 定常状態の近傍を局所的(local)に線形近似で解くアプローチ -->
<!--   - モデルを大域的(global)に解くアプローチ -->


- Dynamic programming can solve the decision-making problem of consumers and firms globally
  - It has its strength when solving the model with strong nonlinearity
  - It can be used for structural estimation
<!-- - 動的計画法は大域的に消費者や企業の意思決定問題を解くことができる -->
<!--   - 強い非線形性が存在するモデルにおいて強みを発揮 -->
<!--   - 構造推計(structural estimation)にも利用される -->

# Robinson Crusoe and Bellman equation
<!-- # ロビンソン・クルーソーとベルマン方程式 -->

- Set up the representative agent model
<!-- - 代表的個人モデル(representative agent model)を設定 -->

- Robinson Crusoe living on a deserted island plans for production and consumption-saving by himself
<!-- - 無人島で生活するロビンソン・クルーソーは、一人で生産と消費・貯蓄の計画を立てる -->

- There is no trade with other economic agents in the market
<!-- - 他の経済主体との市場における取引は存在しない -->

- He plans and does the optimal path of consumption and saving
<!-- - 消費と貯蓄の最適経路(optimal path)を考えて実行する -->

- Time is discrete
  - The current calendar year: $t>0$
  - There will be a help to come after the end of T periods
<!-- - 時間は離散的に進む -->
<!--   - 現時点でのカレンダー年：$t>0$ -->
<!--   - 必ず$T$期終了後に助けが来ることがわかっている：$t=0,...,T$ -->


- At the period of $t=0$, using capital (seeds of crops) $k_{0}$ at hand, he plans the optimal consumption and saving
<!-- - $t=0$期時点で手元にある資本（穀物の種）$k_{0}$を使って、最適な消費・貯蓄計画を立てる -->


- Either to eat the seeds or sow and increase them: If he eats all the seads, there will be nothing to eat in the next period. If he sows all the seads, he is hungry today. Therefore, he must find the best balance
<!-- - この種を食べるか、種を蒔いて増やす：全ての種を食べると次期に食べるものがなくなるが、全ての種を蒔くと今日ひもじい思いをするため、ちょうどよいバランスを考える -->

- Represent this situation as a budget constraint
<!-- - この状況を予算制約として表現すると、 -->

$$
  c_{t} + k_{t+1} \leq f(k_{t}) \tag{1}
$$

- $c_{t}$ is consumption in the period $t$
<!-- - $c_{t}$は$t$期における消費 -->

- $f(k_{t})$ is the production function: If he sows $k_{t}$ units of seeds, he will harvest $y_{t}=f(k_{t})$ seeds
<!-- - $f(k_{t})$は生産関数：$k_{t}$単位の種を蒔くと$y_{t}=f(k_{t})$だけの種が収穫できる -->

- $f'(k_{t})>0$ and $f''(k_{t})<0$ are satisfied (assumption of decreasing returns)
<!-- - $f'(k_{t})>0$および$f''(k_{t})<0$を満たす（収穫逓減の仮定） -->

Robinson Crusoe will maximize the discounted current value of utility from consumption
<!-- - ロビンソン・クルーソーは消費から得られる効用の割引現在価値を最大化する -->

$$
  \max_{\{c_{t},k_{t+1}\}_{t=0}^{T}} \sum_{t=0}^{T} \beta^{t} u(c_{t}) \tag{2}
$$

- Assume $u'(c_{t})>0$ and $u''(c_{t})<0$ for the utiity function $u(c_{t})$
<!-- - 効用関数$u(c_{t})$にも、$u'(c_{t})>0$および$u''(c_{t})<0$を仮定する -->

- Want to know the path of consumption and capital $\{c_{t},k_{t+1}\}_{t=0}^{T}$ so as to maximize the objective function (2)
  - It is optimal to eat all the seeds at period $T$: $k_{T+1}=0$
  - Seeds at hand at period $0$ (he cannot borrow seeds in the deserted island): $k_{0}>0$
<!-- - (2)の目的関数を最大にする消費及び資本の経路$\{c_{t},k_{t+1}\}_{t=0}^{T}$を知りたい -->
<!--   - $T$期にある種は全て食べてしまうのが最適：$k_{T+1}=0$ -->
<!--   - $0$期に手元にある種（無人島では種を借りることはできない）：$k_{0}>0$ -->

- Solve the dynamic optimization problem to maximize (2) subject to the budget constraint (1)
<!-- - 予算制約式(1)のもとで(2)式を最大化する動学的最適化問題を解く -->


- Sequential problem: Using Lagrange method, we can derive the Euler equation as the first order necessary condition
<!-- - 逐次問題(sequential problem)：ラグランジュ未定乗数法を用いると、1階の必要条件であるオイラー方程式を導出できる -->

$$
  u'(c_{t}) = \beta u'(c_{t+1})f'(k_{t+1})
$$

- We will consider another approach here
<!-- - ここでは、別のアプローチを考える -->

- The necessary information for Robinson Crusoe to build the optimal plan includes the length of the periods for which he lives in the deserted island, $T$, and the capital that he owns at the starting period, $k_{0}$
<!-- - ロビンソン・クルーソーが最適計画を立てるのに必要な情報は、無人島で生活する期間$T$とスタート時点で保有している資本$k_{0}$である -->

- Value function: Define $V_{0}(k_{0})$ to be the maximum utility achievable when he owns capital of $k_{0}$ in period 0
<!-- - 価値関数(value function)：$V_{0}(k_{0})$を0期において$k_{0}$だけ資本を保有していた際に達成できる最大効用と定義すると -->

\begin{align*}
  V_{0}(k_{0}) &= \max\{u(c_{0}) + \beta u(c_{1}) + \beta^{2} u(c_{2}) + \cdots + \beta^{T} u(c_{T})\} \\
  & = \max\{u(c_{0}) + \beta [u(c_{1}) + \beta u(c_{2}) + \cdots + \beta^{T-1} u(c_{T})]\} \\
  & = \max\{u(c_{0}) + \beta V_{1}(k_{1})\} \tag{4}
\end{align*}

where $V_{1}(k_{1})$ is the value function when he has capital of $k_{1}$ at period 1
<!-- $V_{1}(k_{1})$は1期に$k_{1}$だけ資本を保有していた際の価値関数 -->

- Even if he owns the same amount of capital in different periods, the value functions are different, $V_{0}(k) \neq V_{1}(k)$, as the remaining durations are different
<!-- - 同じだけ資本を保有していても残存期間が異なれば価値関数は異なる($V_{0}(k) \neq V_{1}(k)$) -->

- The problem for an economic agent who owns $k_{0}$ at period 0 to solve for the optimal saving plan for $T+1$ periods is divided into
  - Consumption-saving choice problem in period 0
  - Optimal saving problem when he owns $k_{1}$ in period 1
<!-- % - 0期に$k_{0}$を保有していた経済主体が$T+1$期間を通じた最適貯蓄計画を解く問題を、 -->
<!--   - 0期の消費・貯蓄選択問題と -->
<!--   - 「1期に$k_{1}$を保有している際の最適貯蓄問題」に分割した -->
  

- Thus, the problem to look for the sequence of consumption and saving for $T+1$ periods is translated into the problem to look for $k_{1}$, taking $V_{1}(k)$ as given
<!-- - これにより、$T+1$期間の消費と貯蓄の流列を探す問題を、（$V_{1}(k)$を所与として）$k_{1}$を探す問題に変換した -->

- Principle of optimality: If the optimal plan is made in period 0, it is also the optimal plan for the sub-problem after that (the optimal saving problem when he owns $k_{1}$ in period 1)
<!-- - 最適性原理(principle of optimality)：0期に最適計画が出来ているのであれば、それ以降の部分問題（「1期に$k_{1}$を保有している際の最適貯蓄問題」）も最適になっている -->


- Eq. (4) is also called Bellman equation
<!-- - (4)式はベルマン方程式(Bellman equation)とも呼ばれる -->

# Solution for the finite period model
## Backward induction
<!-- # 有限期間モデルの解法
## 後ろ向き帰納法
 -->
 
- What we want to know are the value function $V_{t}(k_{t})$ and the policy function $k_{t+1}=g_{t}(k_{t})$
<!-- - 我々が知りたいのは、価値関数$V_{t}(k_{t})$と、政策関数$k_{t+1}=g_{t}(k_{t})$である -->

- As before, we can solve for them backward from the last period
<!-- - 前回同様、最終期から後ろ向きに解いていけばよい -->

- Calibration
  - Robinson Crusoe knows that there will be a help after $T=10$ periods
  - Discount factor: $\beta=0.96$
  - Utility function: $u(c)=\ln c$
  - Production function: $f(k)=k^{\alpha}, \alpha=0.4$
<!-- - カリブレーション -->
<!--   - ロビンソン・クルーソーは、$T=10$期間後に助けが来ることを知っている -->
<!--   - 割引因子：$\beta=0.96$ -->
<!--   - 効用関数：$u(c)=\ln c$ -->
<!--   - 生産関数：$f(k)=k^{\alpha}, \alpha=0.4$ -->

- Discretize the state variable $k$ by the grid for capital $k^{i}\in\{k^{1},...,k^{I}\}$
<!-- - 状態変数$k$を離散化して、$k^{i}\in\{k^{1},...,k^{I}\}$を資本に関するグリッドとする -->

- There is no need to leave capital in the last period $T$
<!-- - 最終期の$T$期では、資本を残しておく必要はない -->

$$
  c^{i}_{T} = f(k^{i}_{T}), \quad V_{T}(k^{i}_{T}) = \ln c^{i}_{T}
$$

- In period $T-1$, the structure is totally the same as in the two-period model we previously studied
<!-- - $T-1$期では、前回勉強した2期間モデルと構造は全く同じである -->

\begin{align*}
  V_{T-1}(k^{i}) &= \max_{k_{T}} u(f(k^{i})-k_{T}) + \beta V_{T}(k_{T}) \\
  &= \max_{k_{T}} \ln(f(k^{i})-k_{T}) + \beta \ln(k_{T})
\end{align*}

- By optimization we introduced before, we obtain an approximated value function (a collection of points)
<!-- - 前回紹介した最適化によって、近似的な価値関数（点の集まり）$\{V_{T-1}(k^{i})\}_{i=1}^{I}$を得られる -->

```julia
# Load packages 
using Dierckx # for interpolation
using Optim
using Plots
```

```julia
# Load modules
include("GenerateGrid.jl")
include("MyEconFcn.jl")
```

```julia
struct Params
    # parameters
    β::Float64 # Discount factor (\beta)
    γ::Float64 # Relative risk aversion (\gamma)
    α::Float64 # capital share (\alpha)
    δ::Float64 # capital depreciation \delta)

    T::Int64 # Years to stay in the deserted island

    # grid
    nk::Int64 # the number of grid points for capital
    kmax::Float64 # the maximum value of the grid for capital
    kmin::Float64 # the minimum value of the grid for capital
    kgrid::Vector{Float64} # the grid for capital
end
```

```julia
function calibration()
    β = 0.96
    γ = 1.0
    α = 0.4
    δ = 1.0 # Not needed?
    T = 10

    nk = 11
    kmax = 1.0
    kmin = 0.05

    # Make equally distanced grid by own code
    kgrid = GenerateGrid.grid_uni(kmin, kmax, nk)
    # Or use it as before
    # kgrid = collect(LinRange(kmin, kmax, nk))

    return Params(β, γ, α, δ, T, nk, kmax, kmin, kgrid)
end
```

```julia
params = calibration();

kmin = 0.05
kmax = 1.0
nk = 11
kgrid = GenerateGrid.grid_uni(kmin, kmax, nk)

# Define variables
pfcn = zeros(params.nk, params.T) # the policy function
vfcn = similar(pfcn) # the value function
cfcn = similar(pfcn) # the consumption function

# Available assets: f(k)
wealth = params.kgrid.^params.α;

# Last period (consume everything)
pfcn[:, end] .= 0.0 # saving is zero as consume everything
cfcn[:, end] = wealth
vfcn[:, end] = MyEconFcn.crra.(cfcn[:, end], params.γ); # the utility from consumption
```

- In period $T-2$, the problem we want to solve is
<!-- % - $T-2$期では、解きたい問題は -->

$$
  V_{T-2}(k^{i}) = \max_{k_{T-1}} u(f(k^{i})-k_{T-1}) + \beta V_{T-1}(k_{T-1}) \tag{5}
$$

- $V_{T-1}(k_{T-1})$ is a collection of the discretized points $\{V_{T-1}(k^{i})\}_{i=1}^{I}$
<!-- - $V_{T-1}(k_{T-1})$は、$\{V_{T-1}(k^{i})\}_{i=1}^{I}$という離散個の点の集まり -->

- There is no guarantee that the optimal capital in the next period will take a value on the grid, therefore we need interpolation
<!-- - 最適な次期の資本$k_{T-1}$が必ずしもグリッド上の値を取る保証はないため、内挿法(interpolation)が必要になる -->

## Interpolation and extrapolation
<!-- ## 内挿法と外挿法 -->

If $k_{T-1}\in[k^{i},k^{i+1}]$ holds,
<!-- $k_{T-1}\in[k^{i},k^{i+1}]$であれば、 -->

- Linear interpolation:
<!-- - 線形補間(linear interpolation)： -->

$$
  V_{T-1}(k_{T-1}) = V_{T-1}(k^{i})\frac{k^{i}-k_{T-1}}{k^{i+1}-k^{i}} + V_{T-1}(k^{i+1})\frac{k_{T-1}-k^{i}}{k^{i+1}-k^{i}}
$$

<!--
- スプライン補間(spline interpolation)：

$$
  V_{T-1}(k_{T-1}) = c_{0,i} + c_{1,i}(k_{T-1}-k^{i}) + c_{2,i}(k_{T-1}-k^{i})^{2} + c_{3,i}(k_{T-1}-k^{i})^{3}   
$$

$(c_{0,i},c_{1,i},c_{2,i},c_{3,i})$は以下の条件を満たすように求められる
-->

- Linear interpolation is intuitive and strong, but there is a drawback
<!-- - 線形補間は直観的で強力であるが、欠点も存在 -->

- Although the value function is differentiable, the approximated function by linear interapolatin is non-differentiable at the knot points between line and line (on the grid)
<!-- - 価値関数は微分可能な関数であるが、それを線形補間して近似した関数は、線と線のつなぎ目（グリッド上）で微分不可能になる -->

- One way to maintain differentiability is to use cubic spline interpolation
<!-- - 微分可能性を維持する方法の1つとして、3次のスプライン補間(cubic spline interpolation)がある -->

```julia
# Interpolation
nkk = 1001
kkmin = params.kmin
kkmax = params.kmax
kkgrid = GenerateGrid.grid_uni(kkmin, kkmax, nkk) #np.linspace(kkmin, kkmax, nkk)

v_ln = zeros(nkk)
v_cs = zeros(nkk)

vfcn1 = vfcn[:,params.T]

interp1_linear = Spline1D(params.kgrid, vfcn1, k=1, bc="extrapolate") # Linear interpolation
interp1_cubic = Spline1D(params.kgrid, vfcn1, k=3, bc="extrapolate") # Spline interpolation

for i = 1:nkk
    v_ln[i] = interp1_linear(kkgrid[i])
    v_cs[i] = interp1_cubic(kkgrid[i])
end
```

- The figure below is the linear approximation and the spline approximation of the value function $V_{T-1}(k^{i})$ in period $T-1$
<!-- - $T-1$期の価値関数$V_{T-1}(k^{i})$のグリッドを線形補間およびスプライン補間したのが以下の図である -->

- When the outside of the domain is complemented, there is a differnt between the linear approximation and the spline approximation
<!-- - 定義域の外側を補完する場合には、線形補間とスプライン補間で違いが出ている -->

```julia
plt = plot(params.kgrid, vfcn[:, end],
    color = :blue,
    marker = :circle,
    label = "data",
    xlims = (0, 1.1),
    ylims = (-1.5, 0.1),
    xlabel = ("Current capital: k"),
    ylabel = ("Value function: V(k)"),
    linewidth = 0,
    markersize = 8,
    markeralpha = 0.5,
    framestyle = :semi
)
plot!(kkgrid,v_ln,color=:blue,label="linear")
plot!(kkgrid,v_cs,color=:orange,label="cubic")
```

## Computation results
<!-- ## 計算結果 -->

- Putting interpolation and optimization together, we can solve Eq. (5) on the grid
<!-- - 内挿法と前回学んだ最適化を組み合わせることで、グリッド上で(5)式を解くことが出来る -->


- We will go backward until period $t=0$ by iterating computation
<!-- - 計算を繰り返し$t=0$期までさかのぼっていく -->

```julia
function BellmanEq(params::Params, wealth::Float64, kprime::Float64, vnext::Spline1D)

    cons = wealth - kprime # consumption level
    current_util = MyEconFcn.crra(cons, params.γ) # utility level
    value = current_util + params.β*vnext(kprime) # u(c)+βV_{t+1}(k')
    value = -1*value # we solve "minimization problem" on the code
    return value

end
```

```julia
# The main loop
for t = params.T-1:-1:1 # Solve from backward with regard to the periods of stay

    # Interpolate the next period's value function
    #vnext = Spline1D(m.kgrid, vfcn[:, t+1], k=1, bc="extrapolate") # Linear interpolation
    vnext = Spline1D(params.kgrid, vfcn[:, t+1], k=3, bc="extrapolate") # Spline interpolation

    for i = 1:params.nk # do loop for the state variable k

        BellmanEq!(kprime) = BellmanEq(params, wealth[i], kprime, vnext)
        res = optimize(BellmanEq!, 0.0, wealth[i], GoldenSection()) # optimization

        pfcn[i, t] = res.minimizer
        vfcn[i, t] = -res.minimum # reverse the sign as we looked for the minimum
    end

    # compute the consumption function
    cfcn[:, t] = params.kgrid.^params.α - pfcn[:, t]
end
```

- The value function is concave, and succeeds the property of the utility function
<!-- - 価値関数は凸関数であり、効用関数の性質を引き継いでいる -->

- There exists the analytical solution for the model with finite $T$ periods
<!-- - 有限$T$期間のモデルには解析的解が存在している -->

- There are some errors with 11 grid points
