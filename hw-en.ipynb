  1. Derive the numerical solution when the state variable $a_{1}$ and the control variable $a_{2}$ are both discrete. (Hint: Compared with the case of $w_{H}=0$, how does the saving function change?)
  2. Derive the numerical solution when the control variable $a_{2}$ is continuous.
